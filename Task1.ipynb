# Example FCD with textual descriptions

```python
fcd_data = [
    {"location": "Highway A", "description": "Heavy rain and strong winds."},
    {"location": "Highway B", "description": "Clear skies and sunny."},
    # More data...
]
```

# Initialize Vader sentiment analyzer

```python
nltk.download('vader_lexicon')
sid = SentimentIntensityAnalyzer()
```

# Analyze sentiment of descriptions

```python
for data_point in fcd_data:
    description = data_point["description"]
    sentiment_score = sid.polarity_scores(description)
    data_point["sentiment_score"] = sentiment_score["compound"]
```

# Print sentiment scores

```python
for data_point in fcd_data:
    print(f"Location: {data_point['location']}, Sentiment Score: {data_point['sentiment_score']}")
```
